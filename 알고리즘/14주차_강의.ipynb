### Bitcoin

1. P2P(Peer to Peer)  
    - Client-Server model  
        1. 내부적, 외부적으로 모두 공격에 취약  
        2. 관리하기가 편리  
    - P2P model  
        1. 공격당해도 큰 영향 없음  


2. Hash code  
    - 해시 함수를 이용해서 입력받은 데이터를 항상 일정한 길이의 문자열로 출력함.
    - 해시 함수 출력값을 이용해 두 입력값의 일치 여부를 판단할 수 있다. 즉 데이터의 변조 여부를 확인할 수도 있다.
    - 해시 함수는 입력값이 같다면 항상 같은 결과를 출력한다. 즉 입력값이 다르면 해시값이 절대로 같을 수 없다.
    - 해시 함수의 출력값으로 입력값을 복호화할 수 없기 때문에 해시 함수는 암호화 방식이 아니다.  
    - 512bit hash algorithm이나 256bit hash algorithm을 주로 사용한다.
    
    
3. Hash cash  
    - 스팸 메일을 필터링할 때 사용하기도 한다.  
    - Bitcoin의 코인 채굴에도 사용하는 방식이다.
    - 본래 입력값은 사실상 찾을 수 없지만, 해시 코드의 일부가 같은 입력값은 비교적 쉽게 찾을 수 있다는 점을 이용한다.
    
    
4. Asymmetric key algorithm(비대칭 키 알고리즘)  
    - 공개키(Public key)와 개인키(Privit key)를 이용한다.
    - 공개키는 네트워크 상에 노출되어 있다.
    - A가 B에게 데이터를 전송할 때, 전송할 데이터는 B의 공개키를 이용해 암호화한다.
    - 그리고 복호화할 때는 B의 개인키를 이용한다.
    - 위의 과정을 이용함으로써 각 사용자의 개인키는 공개하지 않으면서도 암호화된 통신이 가능하다.
