```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
text = open("C:\\Users\\yjy99\\Desktop\\Python\\Python_sync\\파이썬 머슬캠프\\DataFile\\data\\09. alice.txt").read()

alice_m = np.array(Image.open("C:\\Users\\yjy99\\Desktop\\Python\\Python_sync\\파이썬 머슬캠프\\DataFile\\data\\09. alice_mask.png"))

stwd = set(STOPWORDS)
stwd.add("said")
```

```python
import platform
import matplotlib.pyplot as plt

path = "C:\\Windows\\Fonts\\서울한강체 B.ttf"
from matplotlib import font_manager, rc

if platform.system() == "Darwin":
    rc("font", family = "AppleGothic")
    
elif platform.system() == "Windows":
    font_name = font_manager.FontProperties(fname = path).get_name()
    rc("font", family = font_name)
    
else:
    print("Not available system")
    
%matplotlib inline
```

```python
plt.figure(figsize = (8, 8))

plt.imshow(alice_m, cmap = plt.cm.gray, interpolation = "bilinear")

plt.axis("off")

plt.show()
```

```python
wc = WordCloud(background_color = "white", max_words = 2000, mask = alice_m, stopwords = stwd)

wc = wc.generate(text)

wc.words_
```

```python
plt.figure(dpi = 300, figsize = (6, 6))
plt.imshow(wc, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
